```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('online_retail.csv', encoding='unicode_escape')
```

```python
df.isnull().sum()
```

```python
df=df.dropna(subset=['CustomerID'])
```

```python
df.isnull().sum()
```

```python
df.tail(18)
```

```python
df=df[df['Quantity']>0]
```

```python
df.shape
```

```python
user_item_matrix=df.pivot_table(index='CustomerID', columns='Description', values='Quantity',aggfunc='sum', fill_value=0)
```

```python
user_item_matrix.shape
```

```python
item_matrix=user_item_matrix.T
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
raw_similarities=cosine_similarity(item_matrix)
```

```python
item_similarity_df=pd.DataFrame(raw_similarities, 
    index=item_matrix.index, 
    columns=item_matrix.index
)
```

```python
item_similarity_df.shape
```

```python
def get_recommendations(item_name, similarity_matrix, top_n=5):
    
    item_scores = similarity_matrix[item_name]
    sorted_scores = item_scores.sort_values(ascending=False)
    top_items = sorted_scores.iloc[1:top_n+1]
    return top_items


target_item = 'WHITE HANGING HEART T-LIGHT HOLDER'
print(f"Top 5 recommendations for customers who bought: {target_item}\n")


recommendations = get_recommendations(target_item, item_similarity_df)
print(recommendations)
```
